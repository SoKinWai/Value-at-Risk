```python
import yfinance as yf
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd
```

```python
tickers=['AAPL','GOOG','C','DIS','F','MSFT','MS','GME','TSLA','AMZN']
start=dt.datetime(2019,1,1)
end=dt.datetime.now()
```

```python
def getData(stocks, start, end):
    stockData = yf.download(stocks,start,end)['Adj Close']
    returns = stockData.pct_change()
    return returns
```

```python
returns =getData(tickers, start, end)
returns
```

```python
returns = returns.dropna()
returns
```

```python
weights = np.random.random(len(returns.columns))
weights
```

```python
weights = weights/np.sum(weights)
weights
```

```python
#Get the portfolio weighted average return for each day. 
# @ means Matrix Multiplication
returns['Portfolio'] =returns@weights
returns
```

```python
#If you want 90% VaR, you should enter alpha=10
#If you want 99% VaR, you should enter alpha=1
def historicalVaR(returns, alpha=5):
    """
    Read in a pandas dataframe of returns / a pandas series of returns
    Output the percentile of the distribution at the given alpha confidence level
    """
    if isinstance(returns, pd.Series):
        return np.percentile(returns, alpha)
    
   
    elif isinstance(returns, pd.DataFrame):
        
         # A passed user-defined-function will be passed a Series for evaluation.
        return returns.aggregate(historicalVaR, alpha=alpha)
    
    else:
        raise TypeError("Expected returns to be dataframe or series")
```

```python
# 100 days
Time = 100
hVaR=historicalVaR(returns['Portfolio'], alpha=5)*np.sqrt(Time)
hVaR
```

```python
historicalVaR(returns, alpha=5)*np.sqrt(Time)
```

```python
# 90% VaR
historicalVaR(returns, alpha=10)*np.sqrt(Time)
```

```python
# 99% VaR
historicalVaR(returns, alpha=1)*np.sqrt(Time)
```

```python
InitialInvestment = 10000
print('Value at Risk 95th CI    :      ', round(InitialInvestment*-hVaR,2))
```
